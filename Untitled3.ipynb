```python
import torch
import time
import gc
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------
# CONFIG
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# --------------------------------------------
# UTILITY FUNCTIONS

def load_model_original():
    """
    Replace this with actual model loading from the repo.
    """
    model = torch.nn.Sequential(torch.nn.Conv2d(3, 3, 3, padding=1))  # Dummy model
    return model.to(device)

def load_sample_input():
    """
    Replace this with actual preprocessing/input loading logic.
    """
    dummy_input = torch.randn(1, 3, 256, 256).to(device)
    return dummy_input

def run_inference(model, input_tensor):
    return model(input_tensor)

def show_output(output_tensor, title="Output"):
    output_np = output_tensor.squeeze().detach().cpu().numpy().transpose(1, 2, 0)
    plt.imshow(np.clip(output_np, 0, 1))
    plt.title(title)
    plt.axis('off')
    plt.show()

# --------------------------------------------
# ORIGINAL INFERENCE

gc.collect()
torch.cuda.empty_cache()

model_orig = load_model_original()
model_orig.eval()
input_tensor = load_sample_input()

with torch.no_grad():
    start = time.time()
    output_orig = run_inference(model_orig, input_tensor)
    end = time.time()

orig_time = end - start
orig_mem = torch.cuda.memory_allocated(device) / (1024 ** 2)

print(f"\n[Original]")
print(f"Inference Time: {orig_time:.3f} seconds")
print(f"GPU Memory Usage: {orig_mem:.2f} MB")
show_output(output_orig, "Original Output")

# --------------------------------------------
# OPTIMIZED INFERENCE

gc.collect()
torch.cuda.empty_cache()

# Optimization: TorchScript + Mixed Precision
model_opt = torch.jit.script(model_orig)
model_opt.eval()

with torch.no_grad(), torch.cuda.amp.autocast():
    start = time.time()
    output_opt = run_inference(model_opt, input_tensor)
    end = time.time()

opt_time = end - start
opt_mem = torch.cuda.memory_allocated(device) / (1024 ** 2)

print(f"\n[Optimized]")
print(f"Inference Time: {opt_time:.3f} seconds")
print(f"GPU Memory Usage: {opt_mem:.2f} MB")
show_output(output_opt, "Optimized Output")

# --------------------------------------------
# SUMMARY

print("\n=== Performance Summary ===")
print(f"{'Metric':<20} {'Original':<15} {'Optimized'}")
print(f"{'-'*50}")
print(f"{'Inference Time (s)':<20} {orig_time:<15.3f} {opt_time:.3f}")
print(f"{'GPU Memory (MB)':<20} {orig_mem:<15.2f} {opt_mem:.2f}")

# Optional future work:
# - ONNX export and TensorRT benchmarking
# - Model pruning or quantization
# - Layerwise profiling and optimization
```
